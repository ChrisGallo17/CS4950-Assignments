```python
import pandas as pd
import networkx as nx
df = pd.read_csv("rate.csv")
```

```python
df.head()
```

```python
G_panda = nx.from_pandas_edgelist(df, 'From', 'To', edge_attr= 'Rate', create_using = nx.DiGraph())
```

```python
create_using = nx.DiGraph()
nx.shell_layout(G_panda)
```

```python
nx.draw_networkx(G_panda)
```

```python
G_panda.degree()
```

```python
indegreeview = G_panda.in_degree()
```

```python

for employee in indegreeview:
    if 'mostRated' not in locals():
        mostRated = employee[0]
        print (mostRated)
    print(employee[1])
```
