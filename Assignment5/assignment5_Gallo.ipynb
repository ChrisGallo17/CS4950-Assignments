```python
import networkx as nx
import matplotlib.pyplot as plt

G1 = nx.read_edgelist('random1.edgelist')
degrees = dict(G1.degree())
degree_values = sorted(set(degrees.values()))
in_hist = [list(degrees.values()).count(x) for x in degree_values]
print(degree_values, in_hist)
plt.plot(degree_values, in_hist, 'o')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
G2 = nx.read_edgelist('random2.edgelist')
degrees = dict(G2.degree())
degree_values = sorted(set(degrees.values()))
in_hist = [list(degrees.values()).count(x) for x in degree_values]
plt.plot(degree_values, in_hist, 'o')
plt.xlabel('Degree')
plt.ylabel('Fraction of Nodes')
plt.xscale('log')
plt.yscale('log')
plt.show()
print(nx.average_shortest_path_length(G2))
print(nx.average_clustering(G2))
```
